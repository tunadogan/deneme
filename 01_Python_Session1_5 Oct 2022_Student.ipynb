## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Matter of Quotes</p>

**uzun bir text için 3 tırnak (3 çift veya 3 tek) kullanıyoruz**

```python
print("çift tırnakla başladım çift tırnakla bitirdim")
```

```python
print('tek tırnakla başladım tek tırnakla bitirdim')
```

```python
print("Hello World!')
```

```python
""" Burada hata kodunda sonu gelmiyor diyor. Çift tırnak açtın ama kapatmadın.
hala bir çift tırnak bekliyorum diyor. Sondaki tek tırnağı string bir karakter olarak görüyor.
niye? çünkü çift tırnak ile açıldı, kapanması için çift tırnak bekliyor.
artık o sadece çift tırnak ile kapatılabilir.
ikinci çift tırnağı bulana kadar aradaki tüm karakterleri text olarak algılayacak"""
```

```python
print("""benim adım Edwin, 
      ben sizi seviyorum
      sizde huzur buluyorum""")
```

```python
print("""hello world""")
```

```python
print('''hello world''')
```

```python
print("""123456789*-*/=-+.,^&/(+&/|~ef """)
```

```python
"""niye 3 tırnak var peki düşündünüz mü?
...
birden fazla satırın veya uzun metinlerin içinde neler olabilir bir düşünün
içinde tek tırnak olabilir. kesme / olabilir. I am derken I'dan sonra metnin içinde
üstten virgül gelebilir. metinin içinde çift tırnak " kullanmak isteyebilirsin."""
```

```python
print("Ahmet Bey'in kedisi")
```

**cümle içinde tek tırnak hatalı kullanıma örnek**

```python
print('Sevim'in kedisi')
```

```python
print("Sevim'in kedisi")
```

```python
print('Sevim"in kedisi')
```

**hatalı bir tırnak açma kapamaya örnek:**
